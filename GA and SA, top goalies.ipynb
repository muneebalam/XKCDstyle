```python
#min 20k saves, plus Lundqvist
#source: hockey-reference.com
names = ['Brodeur', 'Luongo', 'Roy', 'Joseph', 'Belfour', 
         'Vanbiesbrouck', 'Burke', 'Khabibulin', 'Lundqvist']
sv = [28928, 26739, 25807, 24279, 22434,
     22176, 21009, 20258, 19970]
ga = [2781, 2349, 2546, 2516, 2317,
     2503, 2290, 2071, 1746]
svpct = [1 - g/(g+s) for g, s in zip(ga, sv)]
```

```python
scatter(sv, ga, alpha=0.5, color='lightgreen', s=200)
for i in range(len(names)):
    annotate(names[i], xy=(sv[i], ga[i]), va='center', ha='center')
ylabel('Goals against')
xlabel('Saves')
title('NHL goalies, career regular season, saves vs goals against')
```

```python
scatter(sv, svpct, alpha=0.5, color='lightgreen', s=200)
for i in range(len(names)):
    annotate(names[i], xy=(sv[i], svpct[i]), va='center', ha='center')
ylabel('Save percentage')
xlabel('Saves')
title('NHL goalies, career regular season, saves vs sv pct')
```
