```python
from pylab import *
%matplotlib inline
import matplotlib.cm as mcm
```

```python
xkcd()
rcParams['font.size']=20
fig = gcf()
ax = gca()
xticks([]) # labels
yticks([])
ax.xaxis.set_ticks_position('none') # tick markers
ax.yaxis.set_ticks_position('none')
for side in ['bottom','right','top','left']:
    ax.spines[side].set_visible(False)
    
dps = fig.dpi_scale_trans.inverted()
bbox = ax.get_window_extent().transformed(dps)
width, height = bbox.width, bbox.height

# manual arrowhead width and length
hw = 1./20
hl = 1./20
lw = 1. # axis line width
ohg = 0.3 # arrow overhang

# compute matching arrowhead length and width
yhw = hw * height/width
yhl = hl * width/height

# draw x and y axis
ax.arrow(0, 0, 1, 0., fc='k', ec='k', lw = lw,
         head_width=hw, head_length=hl, overhang = ohg,
         length_includes_head= True, clip_on = False)

ax.arrow(0, 0, 0., 1, fc='k', ec='k', lw = lw,
         head_width=yhw, head_length=yhl, overhang = ohg,
         length_includes_head= True, clip_on = False)

clip_on = True #if only positive x or y values.

xlim(0, 1)
ylim(0, 1)

names = ['Gretzky', 'Hemsky', 'Hall', 'McDavid', 'Pronger', 'Samsonov']
values = [1, 1/11, 0, 0.5, 1, 4]

bar([i/6 for i in range(len(names))], width = 0.12, height=array(values)/4, color = 'r')
xticks([i/6 for i in range(len(names))], names, rotation = 45)

ylabel('Playoffs per year')
yticks(arange(0, 1.1, 0.25), arange(0, 4.1, 1))
```
