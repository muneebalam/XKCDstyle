```python
from pylab import *
%matplotlib inline
import matplotlib.cm as mcm
```

```python
xkcd()
rcParams['font.size']=20
fig = gcf()
ax = gca()
xticks([]) # labels
yticks([])
ax.xaxis.set_ticks_position('none') # tick markers
ax.yaxis.set_ticks_position('none')
for side in ['bottom','right','top','left']:
    ax.spines[side].set_visible(False)
    
dps = fig.dpi_scale_trans.inverted()
bbox = ax.get_window_extent().transformed(dps)
width, height = bbox.width, bbox.height

# manual arrowhead width and length
hw = 1./20
hl = 1./20
lw = 1. # axis line width
ohg = 0.3 # arrow overhang

# compute matching arrowhead length and width
yhw = hw * height/width
yhl = hl * width/height

# draw x and y axis
ax.arrow(0, 0, 1, 0., fc='k', ec='k', lw = lw,
         head_width=hw, head_length=hl, overhang = ohg,
         length_includes_head= True, clip_on = False)

ax.arrow(0, 0, 0., 1, fc='k', ec='k', lw = lw,
         head_width=yhw, head_length=yhl, overhang = ohg,
         length_includes_head= True, clip_on = False)

clip_on = True #if only positive x or y values.

xlim(0, 1)
ylim(0, 1)

xs = [0.9, 0.6, 0.1]
ys = [0.9, 0.7, 0.2]
labels = ['Laura', 'Logan', 'Charles'] 
scatter(xs, ys, color='c', label = 'Logan', s=500, alpha = 0.5)
for i in range(len(xs)):
    annotate(labels[i], xy=(xs[i], ys[i]), va='center', ha='center', fontsize=12)
    
xs = [0.8, 0.7]
ys = [0.3, 0.3]
labels = ['Logan', 'Yukio'] 
scatter(xs, ys, color='g', label = 'The Wolverine', s=500, alpha = 0.5)
for i in range(len(xs)):
    annotate(labels[i], xy=(xs[i], ys[i]), va='center', ha='center', fontsize=12)

xs = [0.6, 0.5, 0.3, 0.2]
ys = [0.4, 0.1, 0.3, 0.4]
labels = ['Raven', 'Logan', 'Eric', 'Charles'] 
scatter(xs, ys, color='r', label = 'Days of\nFuture Past', s=500, alpha = 0.5)
for i in range(len(xs)):
    annotate(labels[i], xy=(xs[i], ys[i]), va='center', ha='center', fontsize=12)

ylabel("Feels")
xlabel('Badassery')
legend(loc=2, bbox_to_anchor=(1, 1), fontsize=12, scatterpoints=1)
title('The Wolverine in movies', size=20)
```
